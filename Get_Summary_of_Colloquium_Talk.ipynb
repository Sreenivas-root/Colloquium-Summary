```python
import google.generativeai as genai
from google.api_core import retry

retry_policy = {"retry": retry.Retry(predicate=retry.if_transient_error)}
```

```python
from google.colab import userdata

GOOGLE_API_KEY = userdata.get('API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
from pydub import AudioSegment
from IPython.display import Audio

full_speech = AudioSegment.from_mp3("a-a.mp3")

# Preview the first 30 seconds.
first_30s_speech = full_speech[:30000]
first_30s_speech
```

```python
uploaded_speech = genai.upload_file(path='a-a.mp3')
```

```python
prompt = "Write a summary of the talk in about 500 words that highlights all the key points presented in the speech"

model = genai.GenerativeModel('gemini-1.5-flash-latest')
response = model.generate_content([prompt, uploaded_speech], request_options=retry_policy)
# print(response.text)
```

```python
from IPython.display import display

display(response.text)
```
